```python
import random
#constant variable
#ASCII Art 만들어주기
HANGMANPICS  = [ '''
  + --- +
  | |
      |
      |
      |
      |
========= ''' , '''
  + --- +
  | |
  O |
      |
      |
      |
========= ''' , '''
  + --- +
  | |
  O |
  | |
      |
      |
========= ''' , '''
  + --- +
  | |
  O |
/ | |
      |
      |
========= ''' , '''
  + --- +
  | |
  O |
 / | \ |
      |
      |
========= ''' , '''
  + --- +
  | |
  O |
 / | \ |
 / |
      |
========= ''' , '''
  + --- +
  | |
  O |
 / | \ |
 / \ |
      |
========= ''' ]

# 맞출 동물 리스트 만들어주기 (split method 활용) #\는 줗바꿈!
words = ('ant baboon badger bat bear beaver camel cat clam cobra \
        cougar coyote crow deer dog donkey duck eagle ferret fox \
        frog goat goose hawk lion lizard llama mole monkey moose \
        mouse mule newtotter owl panda parrot pigeon python rabbit \
        ram rat raven rhino salmon seal shark sheep skunk sloth snake \
        spider stork swan tiger toad trout turkey turtle weasel whale \
        wolf wombat zebra').split()
print(words)
```

```python
def getRandomWord(wordList):
    # 리스트에서 랜덤한 인덱스의 단어를 뽑아주는 함수
    wordIndex = random.randint(0,len(wordList)-1)
    return wordList[wordIndex]

def displayBoard(missedLetter, correctLetter, secretWord):
    print(HANGMANPICS[len(missedLetter)])
    print()
    
    #틀린 글자 표시해주기
    print('Missed letters', end=' ')
    for letter in missedLetter:
        print(letter, end=' ')
    print()
    
    # 맞춘현황 알려주기
    blank = '_' * len(secretWord) #전체 글자 개수만큼 blank설정.
    
    for i in range(len(secretWord)): #blank를 맞은 글자로 바꿔줄 것.
        if secretWord[i] in correctLetter:
            blank = blank[:i]+secretWord[i]+blank[i+1:] #string은 인덱스만 바꾸는 것이 불가능
    for letter in blank:
        print(letter, end=' ')
    print( )
    
def getGuess(alreadyGuessed):
    while True:
        print('Guess a letter.')
        guess=input()
        guess=guess.lower()
        if len(guess)>1:
            print('You need to enter a single letter.')
        elif guess in alreadyGuessed:
            print('You have already guessed this letter')
        elif guess not in 'qwertyuiopasdfghjklzxcvbnm':
            print('please enter a LETTER.')
        else:
            return guess
    

def playAgain():
    print('Do you wanna play the game again?')
    return input().lower().startswith('y') #startwith(~로 시작하는 것 -->True/False)    
```

```python
#시작
print('H A N G M A N')
missedLetter = ''
correctLetter = ''
secretWord = getRandomWord(words)
gameIsDone=False

while True:
    displayBoard(missedLetter, correctLetter, secretWord)
    
    guess = getGuess(missedLetter+correctLetter)
    
    if guess in secretWord: #route1: 추측한 단어가 secretWord 안에 있다면
        correctLetter=correctLetter+guess
    
        foundAllLetters=True #모든 글자를 맞췄는지 확인하기
        for i in range(len(secretWord)):
            if secretWord[i] not in correctLetter:
                foundAllLetters = False
                break
        if foundAllLetters==True:
            print('You won the game! the correct word is '+secretWord)
            gameIsDone=True
    
    else: # route2: 글자가 틀린 경우
        missedLetter=missedLetter+guess
        
        # 게임오버인지 확인
        if len(missedLetter)==len(HANGMANPICS)-1: #행맨 그림의 마지막꺼까지 다 출력직전일 때
            #(이제 마지막꺼 출력하고 꿑내야 함)
            displayBoard(missedLetter, correctLetter, secretWord)
            print('The game is over. The correct word was '+secretWord)
            gameIsDone=True
            
        if gameIsDone==True:
            if playAgain()==True:
                missedLetter=''
                correctLetter=''
                gameIsDone=False
                secretWord=getRandomWord(words)
            else:
                break
            
            
    
```
